# Hospital Readmissions Data Analysis and Recommendations for Reduction

### Background
In October 2012, the US government's Center for Medicare and Medicaid Services (CMS) began reducing Medicare payments for Inpatient Prospective Payment System hospitals with excess readmissions. Excess readmissions are measured by a ratio, by dividing a hospital’s number of “predicted” 30-day readmissions for heart attack, heart failure, and pneumonia by the number that would be “expected,” based on an average hospital with similar patients. A ratio greater than 1 indicates excess readmissions.

B1. Set up an appropriate hypothesis test.

Null Hypothesis: The average Excess Readmission Ratio for hosiptals with less than 100 charges is equal to hospitals with more than 1000 discharges, as well as all hospitals between 100-1000 discharges.

Alternative Hypothesis: The average Excess Readmission Ratio is different between the groups of hospitals.

```python
#I'll start with splitting all the hospitals into groups.
df = clean_hospital_read_df[(clean_hospital_read_df['Excess Readmission Ratio'].notnull())]
low = df[(df['Number of Discharges'] < 100)]
high = df[(df['Number of Discharges'] > 1000)]
mid = df[(df['Number of Discharges'] < 1000) & (df['Number of Discharges'] > 100)]
```

```python
lowd = sorted(low['Excess Readmission Ratio'])
highd = sorted(high['Excess Readmission Ratio'])
midd = sorted(mid['Excess Readmission Ratio'])
```

```python
#This should give us a histogram of the selected data and also show us the normal distribution for it as well.
import scipy.stats as stats
from scipy.stats import f
def normtest(data, title):
    mu = np.average(data)
    sigma = np.std(data)
    fit = stats.norm.pdf(data, mu, sigma)
    plt.plot(data, fit)
    plt.hist(data, 30, normed=True)
    plt.title(title)
    print("n: ", len(data))
    print('Mean: ', mu)
    print('Sigma: ', sigma)
```

```python
#All Data
normtest(sorted(df['Excess Readmission Ratio']), 'test')
```

```python
#Hospitals w/ <100 discharges
normtest(lowd, 'Histogram of Excess Readmission Ratios for Low Discharge Hospitals')
```

```python
#Hospitals w/ >1000 discharges
normtest(highd, 'Histogram of Excess Readmission Ratios for High Discharge Hospitals')
```

```python
#Hospitals w/ 100<x<1000 discharges
normtest(midd, 'Histogram of Excess Readmission Ratios for Mid Discharge Hospitals')
```

Every group of data seems to be close to, if not completely, normally distributed. 

Now to compute the F and P values.

```python
f_val, p_val = stats.f_oneway(lowd, highd, midd)
print(f_val, p_val)
```

With an F-value of 39.64 and a p-value of 6.97e-18, the differences in mean excess readmission ratios among low, mid, and high-discharge hospitals is in fact significant. I would argue, however, that there are still other factors are in play.
